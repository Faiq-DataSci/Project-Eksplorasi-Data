# SINGLE MOVING AVERAGE 3 BULAN

```python
# ==========================================
# IMPORT LIBRARY
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# MEMBACA FILE EXCEL
# ==========================================

df = pd.read_excel(
    "Data.xlsx",
    sheet_name="Produksi_Padi"
)

# print("================ DATA AWAL ================")
# display(df.style.hide(axis="index"))

# ==========================================
# SINGLE MOVING AVERAGE 3 BULAN
# ==========================================

periode = 3

# Nilai St (Forecast)
df["St"] = (
    df["Produksi Padi (Ton) (X2)"]
    .rolling(window=periode)
    .mean()
    .shift(1)
)

# ==========================================
# MENGHITUNG ERROR
# ==========================================

# Error
df["Xt-St"] = df["Produksi Padi (Ton) (X2)"] - df["St"]

# Nilai Absolut Error
df["|Xt-St|"] = df["Xt-St"].abs()

# Kuadrat Error
df["(Xt-St)^2"] = df["Xt-St"] ** 2

# Absolute Percentage Error
df["APE (%)"] = (df["|Xt-St|"] / df["Produksi Padi (Ton) (X2)"]) * 100

print("\n===== HASIL SINGLE MOVING AVERAGE 3 BULAN =====")
display(df.style.hide(axis="index"))

# ==========================================
# MENGHITUNG MAE, MSE, DAN MAPE
# ==========================================

hasil = df.dropna()

mae = hasil["|Xt-St|"].mean()

mse = hasil["(Xt-St)^2"].mean()

mape = hasil["APE (%)"].mean()

print("\n===== HASIL EVALUASI =====")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"MAPE : {mape:.2f}%")

# ==========================================
# VISUALISASI
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(
    df["Bulan"],
    df["Produksi Padi (Ton) (X2)"],
    marker='o',
    linewidth=2,
    label="Data Aktual"
)

plt.plot(
    df["Bulan"],
    df["St"],
    marker='s',
    linewidth=2,
    label="Single Moving Average"
)

plt.title("Single Moving Average Produksi Padi (Ton)")
plt.xlabel("Bulan")
plt.ylabel("Produksi Padi (Ton)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# DOUBLE MOVING AVERAGE

```python
# ==========================================
# IMPORT LIBRARY
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

# ==========================================
# MEMBACA FILE EXCEL
# ==========================================

df = pd.read_excel(
    "Data.xlsx",
    sheet_name="Produksi_Padi"
)

# print("===== DATA AWAL =====")
# display(df.style.hide(axis="index"))

# ==========================================
# DOUBLE MOVING AVERAGE (3 BULAN)
# ==========================================

periode = 3

# Moving Average Pertama (M1)
df["M1"] = (
    df["Produksi Padi (Ton) (X2)"]
    .rolling(window=periode)
    .mean()
)

# Moving Average Kedua (M2)
df["M2"] = (
    df["M1"]
    .rolling(window=periode)
    .mean()
)

# Nilai a
df["a"] = (2 * df["M1"]) - df["M2"]

# Nilai b
df["b"] = (2 / (periode - 1)) * (df["M1"] - df["M2"])

# Forecast periode berikutnya (m = 1)
df["Forecast"] = df["a"] + df["b"]

print("\n===== HASIL DOUBLE MOVING AVERAGE =====")
display(df.style.hide(axis="index"))

# ==========================================
# MENGHITUNG MAE, MSE DAN MAPE 
# ==========================================

hasil = df.dropna()

# Menghitung MAE
mae = mean_absolute_error(
    hasil["Produksi Padi (Ton) (X2)"],
    hasil["Forecast"]
)

# Menghitung MSE
mse = mean_squared_error(
    hasil["Produksi Padi (Ton) (X2)"],
    hasil["Forecast"]
)

# Menghitung MAPE
mape = mean_absolute_percentage_error(
    hasil["Produksi Padi (Ton) (X2)"],
    hasil["Forecast"]
) * 100

print("\n===== HASIL EVALUASI =====")
print(f"MAE : {mae:.2f}")
print(f"MSE : {mse:.2f}")
print(f"MAPE : {mape:.2f}%")

# ==========================================
# VISUALISASI
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(
    df["Bulan"],
    df["Produksi Padi (Ton) (X2)"],
    marker='o',
    linewidth=2,
    label="Data Aktual"
)

plt.plot(
    df["Bulan"],
    df["Forecast"],
    marker='s',
    linewidth=2,
    label="Forecast DMA"
)

plt.title("Double Moving Average Produksi Padi (Ton)")
plt.xlabel("Bulan")
plt.ylabel("Produksi Padi (Ton)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# SINGLE EXPONENTIAL SMOOTHING

```python
# ==========================================
# IMPORT LIBRARY
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

# ==========================================
# MEMBACA FILE EXCEL
# ==========================================

df = pd.read_excel(
    "Data.xlsx",
    sheet_name="Produksi_Padi"
)

# print("===== DATA AWAL =====")
# display(df.style.hide(axis="index"))

# ==========================================
# SINGLE EXPONENTIAL SMOOTHING
# ==========================================

# Mengubah Nilai Alpha (0.10) dan (0.50)
alpha = 0.5

# ==========================================
# PERHITUNGAN SINGLE EXPONENTIAL SMOOTHING
# ==========================================

# Kolom St (hasil smoothing)
df["St"] = 0.0

# Nilai awal St = data aktual pertama
df.loc[0, "St"] = df.loc[0, "Produksi Padi (Ton) (X2)"]

# Menghitung nilai St
for i in range(1, len(df)):
    df.loc[i, "St"] = (
        alpha * df.loc[i-1, "Produksi Padi (Ton) (X2)"]
        + (1 - alpha) * df.loc[i-1, "St"]
    )

# ==========================================
# PERHITUNGAN ERROR
# ==========================================

# Error
df["Xt-St"] = df["Produksi Padi (Ton) (X2)"] - df["St"]

# Nilai Absolut Error
df["|Xt-St|"] = df["Xt-St"].abs()

# Kuadrat Error
df["(Xt-St)^2"] = df["Xt-St"] ** 2

print("\n===== HASIL SINGLE EXPONENTIAL SMOOTHING =====")
display(df.style.hide(axis="index"))

# ==========================================
# MENGHITUNG MAE, MSE, DAN MAPE
# ==========================================

hasil = df.iloc[1:]

mae = hasil["|Xt-St|"].mean()

mse = hasil["(Xt-St)^2"].mean()

mape = (
    (hasil["|Xt-St|"] / hasil["Produksi Padi (Ton) (X2)"])
    .mean()
) * 100

print("\n===== HASIL EVALUASI =====")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"MAPE : {mape:.2f}%")

# ==========================================
# VISUALISASI
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(
    df["Bulan"],
    df["Produksi Padi (Ton) (X2)"],
    marker='o',
    linewidth=2,
    label="Data Aktual"
)

plt.plot(
    df["Bulan"],
    df["St"],
    marker='s',
    linewidth=2,
    label=f"SES (α = {alpha})"
)

plt.title("Single Exponential Smoothing Produksi Padi")
plt.xlabel("Bulan")
plt.ylabel("Produksi Padi (Ton)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

# DOUBLE EXPONENTIAL SMOOTHING

```python
# ==========================================
# IMPORT LIBRARY
# ==========================================
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# MEMBACA FILE EXCEL
# ==========================================

df = pd.read_excel(
    "Data.xlsx",
    sheet_name="Produksi_Padi"
)

# print("================ DATA AWAL ================")
# display(df.style.hide(axis="index"))

# ==========================================
# PARAMETER
# ==========================================

alpha = 0.2

# ==========================================
# MEMBUAT KOLOM
# ==========================================

df["S't"] = 0.0
df["S''t"] = 0.0
df["a"] = 0.0
df["b"] = 0.0
df["Forecast"] = None

# ==========================================
# NILAI AWAL
# ==========================================

df.loc[0, "S't"] = df.loc[0, "Produksi Padi (Ton) (X2)"]
df.loc[0, "S''t"] = df.loc[0, "Produksi Padi (Ton) (X2)"]

# ==========================================
# PERHITUNGAN DOUBLE EXPONENTIAL SMOOTHING
# ==========================================

for i in range(1, len(df)):

    # Single Smoothing
    df.loc[i, "S't"] = (
        alpha * df.loc[i, "Produksi Padi (Ton) (X2)"] +
        (1-alpha) * df.loc[i-1, "S't"]
    )

    # Double Smoothing
    df.loc[i, "S''t"] = (
        alpha * df.loc[i, "S't"] +
        (1-alpha) * df.loc[i-1, "S''t"]
    )

    # Nilai a
    df.loc[i, "a"] = (
        2 * df.loc[i, "S't"] -
        df.loc[i, "S''t"]
    )

    # Nilai b
    df.loc[i, "b"] = (
        alpha / (1-alpha)
    ) * (
        df.loc[i, "S't"] -
        df.loc[i, "S''t"]
    )

    # Forecast satu periode ke depan
    df.loc[i, "Forecast"] = (
        df.loc[i, "a"] +
        df.loc[i, "b"]
    )

# ==========================================
# MENGHITUNG ERROR
# ==========================================

df["Xt-Ft"] = (
    df["Produksi Padi (Ton) (X2)"] -
    df["Forecast"]
)

df["|Xt-Ft|"] = df["Xt-Ft"].abs()

df["(Xt-Ft)^2"] = df["Xt-Ft"]**2

df["APE (%)"] = (
    df["|Xt-Ft|"] /
    df["Produksi Padi (Ton) (X2)"]
) * 100

print("\n===== HASIL DOUBLE EXPONENTIAL SMOOTHING =====")
display(df.style.hide(axis="index"))

# ==========================================
# EVALUASI
# ==========================================

hasil = df.dropna()

mae = hasil["|Xt-Ft|"].mean()

mse = hasil["(Xt-Ft)^2"].mean()

mape = hasil["APE (%)"].mean()

print("\n===== HASIL EVALUASI =====")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"MAPE : {mape:.2f}%")

# ==========================================
# VISUALISASI
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(
    df["Bulan"],
    df["Produksi Padi (Ton) (X2)"],
    marker='o',
    linewidth=2,
    label="Data Aktual"
)

plt.plot(
    df["Bulan"],
    df["Forecast"],
    marker='s',
    linewidth=2,
    label=f"DES Brown (α={alpha})"
)

plt.title("Double Exponential Smoothing Produksi Padi")
plt.xlabel("Bulan")
plt.ylabel("Produksi Padi (Ton)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
